```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('Salary_Data.csv')
data.head()
```

```python
d1=pd.DataFrame({
    "Age":data["Age"],
    "Salary":data["Salary"]
})
d1
```

```python
sns.scatterplot(x="Age",y="Salary",data=d1)
plt.plot()
```

```python
d2=pd.DataFrame({
    "YearsExperience":data["YearsExperience"],
    "Salary":data["Salary"]
})
sns.scatterplot(x="YearsExperience",y="Salary",data=d2)
plt.plot()
```

```python
sns.pairplot(data=data)
plt.plot()
```

```python
sns.heatmap(data=data.corr(),annot=True)
plt.show()
```

check wheteher is there any missing value or not

```python
data.isnull().sum()
```

```python
inp=data.iloc[ : ,0:2]
inp.head()
out=data["Salary"]
```

```python
x_tr,x_tes,y_tr,y_tes=train_test_split(inp,out,train_size=0.2,random_state=44)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_tr,y_tr)
```

```python
model.predict(x_tes)
```

```python
model.score(x_tes,y_tes)*100
```

```python
import numpy as np
model.predict(np.array([[20, 5]]))
```
